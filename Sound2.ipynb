```python
notes = {
    "A" : 220.00,
    "b" : 233.08,
    "B" : 246.94,
    "C" : 261.63,
    "D" : 293.66,
    "E" : 329.63,
    "F" : 349.23,
    "G" : 392.00
}
```

```python
import numpy as np
def play(note, duration, freq, octave=1, kernel=None):
    size = int(duration * freq)
    s = np.zeros(size)
    if note != "-" and note != "_":
        hz = notes[note] * octave
        for i in range(size):
            x = duration / size * i
            s[i] = np.sin(2 * np.pi * hz * x)
    if kernel:
        return s * kernel(size)
    return s
```
